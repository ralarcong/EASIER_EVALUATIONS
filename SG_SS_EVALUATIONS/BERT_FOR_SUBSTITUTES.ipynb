```python
import torch
from transformers import BertForMaskedLM, BertTokenizer
from evaluations import evaluations
import csv
import spacy
tokenizer = BertTokenizer.from_pretrained("pytorch/", do_lower_case=False)
model = BertForMaskedLM.from_pretrained("pytorch/")
evaluationsobj=evaluations()
nlp = spacy.load('es_core_news_md')
model.eval()
```

# Generando sustitutos:

```python
def getsubstitutes(sentence,word):
    sentence = sentence.replace(word, "[MASK]")
    sentence = "[CLS] " + sentence + " [SEP]"
    
    tokens = tokenizer.tokenize(sentence)
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokens)
    tokens_tensor = torch.tensor([indexed_tokens])
    predictions = model(tokens_tensor)[0]
    midx = tokens.index("[MASK]")
    idxs = torch.argsort(predictions[0,midx], descending=True)
    predicted_token = tokenizer.convert_ids_to_tokens(idxs[:3])
    
    return predicted_token
    
```

```python
getsubstitutes('Voy a sacar dinero al banco y luego iré a tomar algo.','banco')
```

```python
def lematizar(word):
    palabra = nlp(word)

    return palabra[0].lemma_

def evaluarbert(path):
    tsvin = open(path, "rt")
    tsvin = csv.reader(tsvin, delimiter=';')
    muestra=list()
    
    for row in tsvin:
        docid = row[0]  # id del parrafo donde ocurre la palabra
        oracion=row[1]
        word=row[2].strip()
        ini=int(row[3])
        fin=int(row[4])
        w1=row[5].strip()
        w2=row[6].strip()
        w3=row[7].strip()
        w4=lematizar(w1)
        w5=lematizar(w2)
        w6=lematizar(w3)
        if w1==w4 and w2==w5 and w3==w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3])
        elif w1==w4 and w2==w5 and w3!=w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w6])
        elif w1==w4 and w2!=w5 and w3==w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w5])
        elif w1!=w4 and w2==w5 and w3==w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w4])
        elif w1==w4 and w2!=w5 and w3!=w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w5,w6])
        elif w1!=w4 and w2==w5 and w3!=w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w4,w6])
        elif w1!=w4 and w2!=w5 and w3==w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w4,w5])
        elif w1!=w4 and w2!=w5 and w3!=w6:
            muestra.append([docid,oracion,ini,fin,word,w1,w2,w3,w4,w5,w6])
        
    return muestra
```

```python
muestrabert=evaluarbert('muestra.csv')
```

```python
result=list()
for i in range(len(muestrabert[:500])):
    result.append(getsubstitutes(muestrabert[i][1],muestrabert[i][4]))
```

# EVALUANDO:

## F1
